```python
import graphviz
```

```python
f = graphviz.Digraph('plain orgonogram 1', format='png')
```

```python
names = ["A","B","C","D","E","F","G","H"]
positions = ["CEO", "Team A Lead", "Team B Lead", "Staff A", "Staff B", "Staff C", "Staff D", "Staff E"]
```

```python
for name,position in zip(names,positions):
    f.node(name,position)
```

```python
f.edge("A","B"); f.edge("A","C")
f.edge("B","D"); f.edge("B","E")
f.edge("C","F"); f.edge("C","G"); f.edge("C","H")
```

```python
f.render(directory='output')
```

```python
f2 = graphviz.Digraph(filename = "output/plain orgonogram 2.gv")
names = ["A","B","C","D","E","F","G","H"]
positions = ["CEO", "Team A Lead", "Team B Lead", "Staff A", "Staff B", "Staff C", "Staff D", "Staff E"]
for name,position in zip(names,positions):
    if name == "A":
        f2.node(name,position,shape="oval")
    elif name in ["B","C"]:
        f2.node(name,position,shape="box")
    else:
        f2.node(name,position,shape="plaintext")
        
f2.edge("A","B"); f2.edge("A","C")
f2.edge("B","D"); f2.edge("B","E")
f2.edge("C","F"); f2.edge("C","G"); f2.edge("C","H")

f2
```

```python
g = graphviz.Graph('questionable',format='png',engine='dot')
word = 'QUESTIONABLE'
nodes = [str(i) for i in range(len(word))]
labels = [ch for ch in word]
for node,label in zip(nodes,labels):
    g.node(node,label)
for i in range(len(word)//2):
    g.edge(nodes[i],nodes[2*i+1])
    if 2*i+2 < len(word):
        g.edge(nodes[i],nodes[2*i+2])
g
        
```
